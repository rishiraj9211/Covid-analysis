```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
covid19_df=pd.read_csv("covid19_india.csv")
df= covid19_df.groupby("Date").Confirmed.sum()
#df.columns["Days","Confirmed"]
date = covid19_df.Date.unique()
#print(df.head())
```

```python
plt.subplot(1, 2, 1)
plt.plot(date,df,linestyle='--')
plt.xlabel("Days")
plt.ylabel("Cases")
plt.title("Covid 19 Analysis")
plt.subplot(1,2,2)
plt.plot(covid19_df["Date"],covid19_df["Deaths"],marker='*')
plt.xlabel("Days")
plt.ylabel("Deaths")
plt.title("Covid 19 Analysis")
plt.subplots_adjust(wspace=.5)
plt.show()
```
